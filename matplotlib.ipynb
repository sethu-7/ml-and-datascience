```python
car=pd.read_csv("car-sales-missing-dropped.csv")
car
```

```python
ts=pd.Series(np.random.random(1000),index=pd.date_range("1/1/2023",periods=1000))
ts=ts.cumsum()
ts.plot()
```

```python
car["Price"]=car["Price"].str.replace("[\$\,\.]","")
car
```

```python
ts=pd.Series(car,index=pd.date_range("1/1/2022"),periods=car.length())
ts=ts.cumsum()
ts.plot()
```

```python
car["total"]=car["Price"].astype(int).cumsum()
car
car.plot(x="sales_date",y="total",kind="bar");
```

```python
car["sales_date"]=pd.date_range("1/1/2022",periods=(len(car)))
```

```python
heart=pd.read_csv("heart-disease.csv")
heart
```

```python
over_50=heart[heart["age"]>50]
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter",x="age",y="chol",c="target")
```

```python
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)

# ax.set_xlim([45,100])
```
